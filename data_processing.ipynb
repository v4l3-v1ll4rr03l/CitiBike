```python
citibike_df = pd.read_csv('Raw Data/JC-201901-citibike-tripdata.csv')
for i in range(len(paths)):
    curr_df = pd.read_csv(paths[i])
    if i > 23:
        curr_df = pd.DataFrame({"starttime":curr_df["started_at"],
                                "stoptime":curr_df["ended_at"],
                                "start station id":curr_df["start_station_id"],
                                "start station name":curr_df["start_station_name"],
                                "start station latitude":curr_df["start_lat"],
                                "start station longitude":curr_df["start_lng"],
                                "end station id":curr_df["end_station_id"],
                                "end station name":curr_df["end_station_name"],
                                "end station latitude":curr_df["end_lat"],
                                "end station longitude" :curr_df["end_lng"]})
    citibike_df = citibike_df.append(curr_df)
citibike_df
citibike_df.to_csv("CitiBike_Data")
```
